# Drought Classification Model Using SPI from Multiple Stations

## Overview
This notebook implements a machine learning approach to classify drought severity 
based on the Standardized Precipitation Index (SPI) calculated from rainfall data 
collected across multiple weather stations.

## Functionality
- Reads and processes rainfall data from multiple station files
- Calculates Standardized Precipitation Index (SPI) values
- Trains and evaluates machine learning models for drought classification
- Visualizes results and model performance

## Dependencies
- pandas, numpy, scikit-learn, matplotlib

```python
#Loading Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import scipy.stats as st

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler
import os

sns.set(style='whitegrid')
```

## ðŸ“¥ Load and Clean Data

## Data Preprocessing  
- Loading rainfall data  
- Cleaning missing values  
- Generating SPI values  
- Normalizing data for LSTM input  


```python
#Loading and clean dataset from csv files

file_paths = [
    './data/Gauge 1 _ Livingstone.csv',
    './data/UmeRiver.csv',
    './data/Gauge 2_ Kalomo.csv',
    './data/Gauge 3_Gwayi River.csv',
]

def clean_station_data(file_path):
    df = pd.read_csv(file_path)
    df.columns = df.columns.str.strip().str.replace('\t', '', regex=False)
    df = df.rename(columns={df.columns[0]: 'Date', df.columns[1]: 'Rainfall'})
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df = df[['Date', 'Rainfall']].dropna()     #dropping empty cells
    df = df.set_index('Date').resample('M').sum()
    df['Station'] = os.path.splitext(os.path.basename(file_path))[0]
    return df

stations = [clean_station_data(f) for f in file_paths]
data = pd.concat(stations).reset_index()
data.head()
```

## ðŸ“ˆ Calculate Standardized Precipitation Index (SPI)

## **Standardized Precipitation Index (SPI) Calculation**
The Standardized Precipitation Index (SPI) quantifies meteorological drought by comparing observed precipitation levels to historical norms. It is computed by normalizing precipitation data into a standard normal distribution.

### **Calculation Steps**
1. **Rolling Summation**: Aggregate precipitation over a defined time window (e.g., 3 months).
2. **Fit a Gamma Distribution**: Monthly precipitation values typically follow a gamma distribution due to their positive, skewed nature.
3. **Compute Cumulative Probability (CDF)**: Convert precipitation values to probabilities using the fitted gamma distribution.
4. **Standard Normal Transformation**: Convert cumulative probabilities into SPI values using the inverse of the normal distribution.
5. **Interpretation**:
   - SPI > -0.5 â†’ No drought (normal conditions)
   - SPI between -0.5 and -0.99 â†’ Mild drought
   - SPI between -1.0 and -1.49 â†’ Moderate drought
   - SPI between -1.5 and -1.99 â†’ Severe drought
   - SPI < -2.0 â†’ Extreme drought

Note : Precipitation data in many regions is likely to contain zeros especially during dry seasons
Fix:
1. Added detection of zero values in the rainfall data and calculated their probability (q)
2. Separated positive values for gamma distribution fitting (since gamma distribution requires strictly positive values)
3. Implemented a mixed distribution approach that accounts for both zero and non-zero values:
   - For zero values: Used the probability q
   - For positive values: Used q + (1-q) * gamma_cdf(value)
4. Added a check to handle cases where there might not be enough positive data points for fitting

This approach follows the standard meteorological practice for SPI calculation when dealing with precipitation data that contains zeros, which is common in many regions, especially during dry seasons.

```python
# Function to calculate SPI using gamma distribution
def calculate_spi_gamma(series, scale=3):
    """
    series (pd.Series): Monthly precipitation data.
    scale (int): Rolling window size for aggregation (default is 3 months).
    """
    rolled = series.rolling(window=scale).sum().dropna()
    
    # Handle zeros in the data (common in precipitation data)
    # Calculate probability of zero values
    q = len(rolled[rolled == 0]) / len(rolled)
    
    # Remove zeros for gamma fitting (gamma distribution requires positive values)
    rolled_positive = rolled[rolled > 0]
    
    if len(rolled_positive) < 2:  # Not enough data for fitting
        return pd.Series(np.nan, index=rolled.index)
    
    # Fit gamma distribution to non-zero precipitation data
    shape, loc, scale_param = st.gamma.fit(rolled_positive, floc=0)
    
    # Compute CDF values, accounting for zeros
    cdf_values = np.zeros(len(rolled))
    
    for i, value in enumerate(rolled.values):
        if value == 0:
            cdf_values[i] = q  # Probability of zero
        else:
            # Adjust probability: P(X â‰¤ x) = q + (1-q) * F(x)
            # where F(x) is the gamma CDF for positive values
            cdf_values[i] = q + (1-q) * st.gamma.cdf(value, shape, loc=loc, scale=scale_param)
    
    # Convert to standard normal distribution (SPI values)
    spi_values = st.norm.ppf(cdf_values)
    
    return pd.Series(spi_values, index=rolled.index)

# Function to classify drought levels based on SPI thresholds
def classify_drought(spi):
    """
    Parameters:
    spi (float): SPI value.
    """
    if spi > -0.5:
        return 'No Drought'
    elif -0.99 < spi <= -0.5:
        return 'Mild Drought'
    elif -1.49 < spi <= -1.0:
        return 'Moderate Drought'
    elif -1.99 < spi <= -1.5:
        return 'Severe Drought'
    else:
        return 'Extreme Drought'

# Apply SPI calculation and drought classification per station
processed = []
for station, group in data.groupby('Station'):
    group = group.copy()
    group['SPI3'] = calculate_spi_gamma(group['Rainfall'])
    group.dropna(inplace=True)
    group['Drought_Class'] = group['SPI3'].apply(classify_drought)
    group['Drought_Label'] = group['Drought_Class'].astype('category').cat.codes

    # Generate lag features for previous SPI values
    for i in range(1, 4):
        group[f'SPI3_t-{i}'] = group['SPI3'].shift(i)
    
    group.dropna(inplace=True)
    processed.append(group)

# Concatenate all processed station data
final_df = pd.concat(processed)
final_df.head()
```

## ðŸ§  LSTM Model: Train/Test Split and Forecasting

```python
# Train/Test Split, Train Model, Evaluate

models = {}
for station, group in final_df.groupby('Station'):
    X = group[[f'SPI3_t-{i}' for i in range(1, 4)]]
    y = group['Drought_Label']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

    clf = RandomForestClassifier(n_estimators=100, random_state=42)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    models[station] = (clf, X_test, y_test, y_pred)

    print(f"\n===== {station} =====")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
```

### âš™ï¸ Data Normalization and Sequence Preparation

```python
# Normalize SPI values for better convergence
scaler = MinMaxScaler()
final_df[['SPI3', 'SPI3_t-1', 'SPI3_t-2', 'SPI3_t-3']] = scaler.fit_transform(final_df[['SPI3', 'SPI3_t-1', 'SPI3_t-2', 'SPI3_t-3']])

# Prepare LSTM-compatible input format
def create_sequences(data, target, sequence_length=3):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i + sequence_length].values)
        y.append(target.iloc[i + sequence_length])
    return np.array(X), np.array(y)

models = {}
sequence_length = 3

for station, group in final_df.groupby('Station'):
    X = group[['SPI3_t-1', 'SPI3_t-2', 'SPI3_t-3']]
    y = group['SPI3']

    # Create sequences
    X_seq, y_seq = create_sequences(X, y, sequence_length)

    # Train/test split
    train_size = int(len(X_seq) * 0.8)
    X_train, X_test = X_seq[:train_size], X_seq[train_size:]
    y_train, y_test = y_seq[:train_size], y_seq[train_size:]

    # Build LSTM model
    model = Sequential([
        LSTM(50, activation='relu', return_sequences=False, input_shape=(sequence_length, X_train.shape[-1])),
        Dense(1)  # Output future SPI value
    ])

    model.compile(optimizer=Adam(learning_rate=0.01), loss='mse')

    # Train LSTM model
    model.fit(X_train, y_train, epochs=50, batch_size=16, verbose=1)

    # Predict next SPI values
    y_pred = model.predict(X_test)

    models[station] = (model, X_test, y_test, y_pred)

    # Plot actual vs predicted SPI values
    plt.figure(figsize=(8, 4))
    plt.plot(y_test, label='Actual SPI', marker='o')
    plt.plot(y_pred, label='Predicted SPI', marker='o')
    plt.legend()
    plt.title(f'Multi-Step SPI Forecasting for {station}')
    plt.xlabel('Time')
    plt.ylabel('SPI Value')
    plt.grid(True)
    plt.show()
```

### ðŸ“ˆ Forecasting Results: Model Evaluation and Visualization

```python
# Save the trained model for each station
for station, (model, X_test, y_test, y_pred) in models.items():
    model.save(f'./models/lstm_spi_{station}.h5')  # Save as .h5 format
    print(f"Model for {station} saved successfully!")
```
